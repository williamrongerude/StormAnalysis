### Impute Missing Values

Magnitude has many nan values, which we could potentially replace with the median, mean, or mode of magnitude. We could also use a specific value based on what results in high metrics. Furthermore, it might be the case that we should drop all rows where magnitude is nan. This is an example of potential modifications where you should try them all, committing to the best change after the validation set.

This is a conventional rule to use quickly for your feature:

If you have enough data where the feature isn't nan, you can drop the rows where the feature is nan. If you're scarce on data, keep the nan rows and find a value to impute.

Lets explore both options.

#### Finding a mean, median or mode value to impute with.

We can inspect magnitude to determine the best decision to make. Lets make a histogram to see the frequency distribution of magnitude.

```python
import matplotlib.pyplot as plt

plt.hist(df['MAGNITUDE'].dropna(), bins=40)
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.title('Magnitude Distribution')
plt.show()

# print median, mean and mode
```

> About 9000 data points have a magnitude value of roughly zero, and 8700 data points have a magnitude value of 50.

> Furthermore, we can note the distribution of our magnitude frequencies and the behavior of the core cluster.

The goal is to find a imputation value that keeps the distribution the same before and after imputation. We can make three dataframe copies, impute the nan with median, mean and mode for all three, and replot the dataframe copies to see which is most similar to our original dataframe. What we want to see is an identical core cluster around the middle peak, meaning the shape of our core is identical to the shape of our original plot's core.

```python
mean = df['MAGNITUDE'].mean()
print('Mean:', mean)

mean_imputed = df['MAGNITUDE'].copy()
mean_imputed.fillna(mean, inplace=True)

plt.figure(figsize=(6.5, 4.5))
plt.hist(mean_imputed, bins=40)
plt.title('Mean Imputation')
plt.show()
```

```python
median = df['MAGNITUDE'].median()
print('Median:', median)

median_imputed = df['MAGNITUDE'].copy()
median_imputed.fillna(median, inplace=True)

plt.figure(figsize=(6.5, 4.5))
plt.hist(median_imputed, bins=40)
plt.title('Median Imputation')
plt.show()
```

```python
mode = df['MAGNITUDE'].mode()[0]
print('Mode:', mode)

mode_imputed = df['MAGNITUDE'].copy()
mode_imputed.fillna(mode, inplace=True)

plt.figure(figsize=(6.5, 4.5))
plt.hist(mode_imputed, bins=40)
plt.title('Mode Imputation')
plt.show()
```

Imputing with the median or mode is ideal here. The core cluster is almost the same as the original plot. The mean plot is the wrong choice. It contains a middle peak around 40, resulting in empty space to the right of the peak, which causes a differently shaped core cluster.

You may be confused about the imbalance between zeros and the middle peak. This is expected. When accounting for nan rows, we are forced to introduce non-zero rows, thereby increasing the middle peak. All we care about is the core cluster's shape, and how much it's preserved after imputation.

Read the second part, where we remove nan-valued rows entirely. Though between that method and imputation, we end up choosing to impute with the median.

```python
median = df['MAGNITUDE'].median()
print('Median:', median)

df['MAGNITUDE'].fillna(median, inplace=True)

plt.figure(figsize=(6.5, 4.5))
plt.hist(df['MAGNITUDE'], bins=40)
plt.title('Median Imputation')
plt.show()
```

#### Removing NAN-valued rows entirely

As we stated, we could also drop nan magnitude rows entirely. We must first check the balance between nan and non-nan rows, to help us determine if we're low on non-nan rows.

This code checks the amount of rows where magnitude is nan and not nan.

```python
nan_count = df['MAGNITUDE'].isna().sum()
non_nan_count = df['MAGNITUDE'].notna().sum()

print("Nan magnitude values:", nan_count)
print("Non-nan magnitude values:", non_nan_count)
```

Almost half of our data would be lost if we removed nan-magnitude rows entirely.

Imagine that magnitude is truly an important feature. Removing nan-magnitude rows could prove to be beneficial, and would also be effectively reducing our learnable space. If we removed the rows, the model's magnitude signal would be truly representative of the real magnitude values, which would result in more accurate magnitude signal, less injected noise from imputation, and *potentially* higher metrics if the magnitude to damaged property relationship is highly influential.

Alternatively, imagine a scenario where other features have more predictive power over magnitude, and imagine the nan-magnitude rows you want to drop also contain these features, such as event_type, tor_width and event_narrative. If these alternative features contain values that represent healthy signal to the model, nan-magnitude imputation would inject numbers that are theoretically noise, but would be noise that is insignificant to the healthy signal from other columns that exist in the rows you would be dropping. Thus, by choosing imputation, the nan issue is solved, and we preserve healthy signal that comes from the rows (but specifically the alternative columns) that we thought about dropping.

Either the signal of magnitude is better, or the signal of other columns (that exist within the rows-to-be-dropped) are better.

TDLR:
> If our feature is a key predictor, we should drop rows where the feature is nan, given that our feature has important signal that we need to model accurately.

> If other features prove to be more beneficial than our feature, then we want to keep as many rows as possible to represent those features. Keep the nan-valued rows and impute the nan values with a working value.

Deciding to impute with a number or drop nan rows entirely should be made based on which provides better metrics in validation.

##### Typecasting

Something to note is that essentialy all model archiectures require integer/float columns. This includes DAMAGE_PROPERTY, which needs to contain floats for this reason and also so we can perform operations on it.

The following code is based on how damaged property is formatted here (K, M, B Notation). We use value.endswith() to check whether the last character is a K, M or B. After finding the true statement, we index the string value from the beginning character to the char before the last, we typecast this substring to a float, and then we multiply the float by the proper constant.

```python
def parse_damage(value):
    if isinstance(value, str):
        value = value.upper().strip()
        if value.endswith('K'):
            return float(value[:-1]) * 1_000
        elif value.endswith('M'):
            return float(value[:-1]) * 1_000_000
        elif value.endswith('B'):
            return float(value[:-1]) * 1_000_000_000
        else:
            return float(value)
    else:
        return value

df['DAMAGE_PROPERTY'] = df['DAMAGE_PROPERTY'].apply(parse_damage)
```

#### Interaction Terms

We can create new feature columns, either by combining features, merging data that's provably causing output change, or mathematically transforming columns to better represent relationships that the model must learn.

Lets create a DURATION_X_MAGNITUDE column. This may expose the combination of event duration and the strength throughout the duration.

Create duration column with time column(s):

```python
df['BEGIN_MINUTES'] = (df['BEGIN_TIME'] // 100) * 60 + (df['BEGIN_TIME'] % 100)
df['END_MINUTES']   = (df['END_TIME'] // 100) * 60 + (df['END_TIME'] % 100)

df['DURATION'] = df['END_MINUTES'] - df['BEGIN_MINUTES']
df.loc[df['DURATION'] < 0, 'DURATION'] += 1440
```

New interaction term:

```python
df['DURATION_X_MAGNITUDE'] = df['DURATION'] * df['MAGNITUDE']
```

```python
df[['DURATION_X_MAGNITUDE', 'DAMAGE_PROPERTY']]
```

Not a bad idea at all. The values are numerically similar in range, which improves the likelihood that the model picks up patterns between them. Try to find more patterns could produce new findings in combination. A column that makes sense to do so, and or a column that just looks sort of like the output column may work.

#### Feature Expansions

Applying transformations like powers, roots, or logarithms to features can reveal nonlinear relationships the model might not detect otherwise. It may help to first make intput and output columns share the same range.

In many situations, applying a log transformation on the output colummn will improve R^2 scores. Log reduces huge ranges, makes skewed data look closer to a normal distribution, and reduces the outlier impact. All of this results in the model fitting a simpler, smoother curve through the data.

```python
import numpy as np

damage_df = df[['DAMAGE_PROPERTY']].copy()
damage_df['LOG_DAMAGE_PROPERTY'] = np.log1p(damage_df['DAMAGE_PROPERTY'])
damage_df
```

Log_damage_property's range is greatly reduced. Keep the original damage property up until train/val/test split so we can preserve an original copy.
